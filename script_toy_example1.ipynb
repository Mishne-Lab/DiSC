```python
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.patches as pathces
import helper_functions as utils
import datasets.toy_example as dataset
# import plot_functions as plots
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
def plot_vectors_combined_n(U,n,num_vecs=2,ymin=None,ymax=None,title=None,figsize=None,savefig=False,save_filename=None):
    dict_ = {0:'A',1:'B',2:'C',3:'D'}
    colors = ['tab:blue','tab:orange']
    markers = ['-','--']
    num_data = len(U)
    cmap = cm.get_cmap(name='rainbow')
    color_vec = cmap(np.arange(256))
    figsize = (12,10) if figsize is None else figsize
    fig,ax = plt.subplots(1,num_data,figsize = figsize)
    for i in range(num_vecs): 
        V,y_i,l_i = [],[],[]
        for j in range(num_data):
            Vcur = U[j][i,:]
#             ax[j].plot(Vcur)
            ax[j].plot(Vcur,markers[i],label="$(V_{})_{}$".format(dict_[j],i+1),color=colors[j],alpha=1-0.4*i)
#             print(colors[i])
            y_i_cur = np.min(Vcur)
            l_i_cur = np.max(Vcur)-np.min(Vcur)
            y_i.append(y_i_cur)
            l_i.append(l_i_cur)
    if ymin is not None and ymax is not None:
        ax[0].set_ylim([ymin,ymax])
        ax[1].set_ylim([ymin,ymax])
        l_i = [ymax-ymin for j in range(num_data)]
        y_i = [ymin for j in range(num_data)]
        
    for n_idx,j in enumerate(n):
        x_j = np.sum(n[:n_idx]) if n_idx else 0
        p_d = n[n_idx]
#         print(x_j,p_d)
        for k in range(num_data):
            rect_cur_j = pathces.Rectangle((x_j,y_i[k]), p_d, l_i[k], color=color_vec[(250//len(n))*n_idx],alpha = 0.1)
            ax[k].add_patch(rect_cur_j)
    plt.suptitle(title,fontsize=20)   
    for i in range(num_data):
        ax[i].legend(fontsize=15,loc = 'upper left')
        ax[i].tick_params(labelsize=18)
        if i:
            ax[i].set_yticks([])
    fig.subplots_adjust(wspace=0.05, hspace=0)
    if savefig:
        plt.savefig(save_filename)
    
```

```python
def plot_vectors(U_A,U_B,n,num_vecs=2,ymin=None,ymax=None,title=None,figsize=None,savefig=False,save_filename=None):
    cmap = cm.get_cmap(name='rainbow')
    color_vec = cmap(np.arange(256))
    colors = ['tab:blue','tab:orange']
    figsize = (12,10) if figsize is None else figsize
    fig,ax = plt.subplots(2,num_vecs,figsize = figsize)
    for i in range(num_vecs):   
        V1 = U_A[i,:]
        V2 = U_B[i,:]
        ax[0,i].plot(V1,color="tab:blue")
        ax[1,i].plot(V2,color="tab:orange")
        ax[0,i].text(0.5, 0.24, "$(U_A)_{}$".format(i+1), bbox=dict(fill=False, edgecolor='k', linewidth=3,alpha=0.0),fontsize=20)
        ax[1,i].text(0.5, 0.24, "$(U_B)_{}$".format(i+1), bbox=dict(fill=False, edgecolor='k', linewidth=3,alpha=0.0),fontsize=20)
    
    #     ax[0,i].set_title("r_k: {}".format(r_k_A_new[non_redun_idx_A[i]]))
    #     ax[1,i].set_title("r_k: {}".format(r_k_B_new[non_redun_idx_B[i]]))
        y_A_i = np.min(V1)
        y_B_i = np.min(V2)
        l_A_i = np.max(V1)-np.min(V1)
        l_B_i = np.max(V2)-np.min(V2)  
        if ymin is not None and ymax is not None:
            ax[0,i-1].set_ylim([ymin,ymax])
            ax[1,i-1].set_ylim([ymin,ymax])
            l_A_i = ymax-ymin
            l_B_i = ymax-ymin
            y_A_i = ymin
            y_B_i = ymin
        for n_idx,j in enumerate(n):
            x_j = np.sum(n[:n_idx]) if n_idx else 0
            p_d = n[n_idx]
    #         print(x_j,p_d)
            rect_A_j = pathces.Rectangle((x_j,y_A_i), p_d, l_A_i, color=color_vec[(250//len(n))*n_idx],alpha = 0.1)
            ax[0,i].add_patch(rect_A_j)
            rect_B_j = pathces.Rectangle((x_j,y_B_i), p_d, l_B_i, color=color_vec[(250//len(n))*n_idx],alpha = 0.1)
            ax[1,i].add_patch(rect_B_j)
    plt.suptitle(title,fontsize=20) 
    for i in range(num_vecs):
        ax[0,i].tick_params(labelsize=18)
        ax[1,i].tick_params(labelsize=18)
        ax[0,i].set_xticks([])
        
        if i:
            ax[0,i].set_yticks([])
            ax[1,i].set_yticks([])
    
    fig.subplots_adjust(wspace=0.05, hspace=0.05)
    if savefig:
        plt.savefig(save_filename)
```

```python
# parameters to generate data
m = 10000
# Toy Example - Case 1
n_A = [100,25,25,25,25,25,25]
type_A = [0,2,2,2,2,0,0]
cov_A = None
n_B = [100,25,25,25,25,25,25]
type_B = [0,1,1,0,0,2,2]

n = [100,25,25,25,25,25,25]

p = np.sum(n_A)
pos_corr = True

# fixing random seed
seed1 = 1010
seed2 = 1111
mul_fac = 0.005
X_A,cov_A = dataset.generate_simulated_data_cases(n_A,type_A,m,cov=cov_A,pos_corr=pos_corr,seed=seed1,mul_fac=mul_fac)
X_B,cov_B = dataset.generate_simulated_data_cases(n_B,type_B,m,cov=cov_A,pos_corr=pos_corr,seed=seed2,mul_fac=mul_fac)

k = 30
d = 20 
results = utils.compute_diff_vectors(X_A.T,X_B.T,k,d)
```
